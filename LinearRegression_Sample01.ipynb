```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Simple example data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)

# Ordinary Least Squares fit
m, b = np.polyfit(x, y, 1)
y_pred = m * x + b

# Mean Squared Error
mse = np.mean((y - y_pred) ** 2)

# Plot
plt.figure(figsize=(7,5))
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred, label="Recta OLS")
for xi, yi, yp in zip(x, y, y_pred):
    plt.vlines(xi, min(yi, yp), max(yi, yp), linestyles='dashed')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresión lineal: OLS y error")
plt.legend()
plt.tight_layout()

plot_path = "/mnt/data/regresion_ols_mse.png"

# Create the directory if it does not exist
os.makedirs(os.path.dirname(plot_path), exist_ok=True)

plt.savefig(plot_path, dpi=150)
plt.show()

print(f"Pendiente (m): {m:.3f}")
print(f"Intercepto (b): {b:.3f}")
print(f"MSE: {mse:.3f}")
print(f"Archivo: {plot_path}")
```
